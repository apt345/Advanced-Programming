Next, you can see the *getMessage()* function. It opens a file that contains a message, reads the message, closes the file, and returns the message as a string.

```python
def getMessage(fileName):
    # Now, we open the file for reading
    myFile = open(fileName, "r")
    # read the complete message.
    message = myFile.read()
    # And close the file
    myFile.close()    
    return(message)
```

(1) (0.25 points) Program a function that takes a message (a string) as input and returns a list with the words in the string. Something like this:

**CODE:**

message = 'In a hole in the ground, there lived a Hobbit'

print(returnWords(message))

**RESULT:**

['in', 'a', 'hole', 'in', 'the', 'ground', 'there', 'lived', 'a', 'hobbit']

```python
#we have to take into account all the possible separators
def splitting(message):
    #add all the separators symbols
    from string import punctuation
    separators=set("\u005c")|set(punctuation)|set(" ") #cleaner 
    separators=list(separators) #transform it back to a list
    
    for sep in separators:
        #print(sep)
        if sep in message:
            message=message.replace(sep, ",")
            message=message.replace(",,",",")#eliminate duplicated commas (blankspace comma or similar)
            message=message.replace(",,", ",")#eliminate 3,4 commas 
            #print(message)
    final=message.split(",")
    #check if the last and first item is empty: phrase ending in . or ?! and remove it
    if final[len(final)-1]=="":
        del(final[len(final)-1])
    if final[0]=="":
        del(final[0])
        
    return(final)

#check that it is working properly
splitting('in a hole in the ground, there lived a Hobbit')
splitting('a very complicated phrase: with different separators; many \ words! And who knows what more?')
```

(2) (0.25 points) Define a function, that returns a dictionary, that counts the words in a sentence.

**CODE**

words = ['in', 'a', 'hole', 'in', 'the', 'ground', 'there', 'lived', 'a', 'hobbit']

counts = countWords(words)

print(list(counts.items()))

**RESULT**

[('in', 2), ('a', 2), ('hole', 1), ('the', 1), ('ground', 1), ('there', 1), ('lived', 1), ('hobbit', 1)]

```python
from collections import defaultdict

def countWords(sentence):
    my_dict=defaultdict(lambda: 0)

    keys=splitting(sentence.lower())#split in words in lower case so that we can count them

    for key in keys:
        my_dict[key]+=1#count them, using default dictionaries, when going for a non present key, it starts at 0

    return(my_dict)
    
countWords('In a hole in the ground, there \\\ \\lived a Hobbit')

```

(4) (0.25 points) Now, compute the set of **unique** words and sort them.

- In order to get the unique word of a message, you can use sets. That is, if you transform each message to a set, you will get the unique words in that message. 
- In order to compute the unique words of **all** messages, you can use the union of sets.

```python
unique=[]
totalunique=set("")
for i in range(0, len(results)):
    unique.append(set(results[i]))
for j in range(1, len(unique)):
    totalunique=unique[j]|unique[j-1]|totalunique
totalunique=sorted(totalunique)
#print(totalunique)
```
